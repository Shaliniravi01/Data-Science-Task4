```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train=pd.read_csv("twitter_training (1).csv")
```

```python
train.dropna(inplace=True)
```

```python
train.isnull().sum()
```

```python
plt.figure(figsize=(8,10))
train['Positive'].value_counts().plot(kind='barh',color='orange')
plt.xlabel("Count")
plt.show()
```

```python
sentiment_counts = train['Positive'].value_counts()

# Create the pie chart
plt.figure(figsize=(5, 8))
colors=['red','pink','blue','yellow']
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct="%1.1f%%", startangle=140,colors=colors)

plt.title('Sentiment Distribution')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(20,12))
sns.countplot(x='Borderlands',data=train,palette='viridis')
plt.xticks(rotation=90)
plt.show()
```

```python
# Create the crosstab
crosstab = pd.crosstab(index=train['Positive'], columns=train['Borderlands'])

# Plot the heatmap
plt.figure(figsize=(20, 8))  
sns.heatmap(crosstab, cmap='coolwarm', annot=True, fmt='d', linewidths=.5)

# Add labels and title
plt.title('Heatmap of Positive vs Borderlands')
plt.xlabel('Positive')
plt.ylabel('Borderlands')

# Show the plot
plt.show()
```
